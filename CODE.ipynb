```python
import glob
import cv2

path = glob.glob("assets/*.jpeg")

for index, file in enumerate(path):
    img = cv2.imread(file)
    cv2.imshow(f'Image {index + 1}', img)

cv2.waitKey(0)          # menunggu sampai user menekan key pada keyboard
cv2.destroyAllWindows() # menutup semua opencv windows
```

```python
img1 = cv2.imread("assets/Anya.jpeg")
img2 = cv2.imread("assets/Anya_2.jpeg")

result = cv2.add(img1, img2)

cv2.imshow("Gambar 1", img1)
cv2.imshow("Gambar 2", img2)
cv2.imshow("Nambahin Gambar", result)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
img1 = cv2.imread("assets/Anya.jpeg")
img2 = cv2.imread("assets/Anya_2.jpeg")

result = cv2.subtract(img1, img2)

cv2.imshow("Gambar 1", img1)
cv2.imshow("Gambar 2", img2)
cv2.imshow("Ngurangin Gambar", result)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
img1 = cv2.imread("assets/Anya.jpeg")
img2 = cv2.imread("assets/Anya_2.jpeg")

result = cv2.addWeighted(img1, 0.9, img2, 0.1, 100)

cv2.imshow("Gambar 1", img1)
cv2.imshow("Gambar 2", img2)
cv2.imshow("Tambah dengan Bobot", result)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
img = cv2.imread("assets/Ferrari.jpg")

# Calculate RGB Channel Histogram
histrBlue = cv2.calcHist([img], [0], None, [256], [0, 256])
histrGreen = cv2.calcHist([img], [1], None, [256], [0, 256])
histrRed = cv2.calcHist([img], [2], None, [256], [0, 256])

import matplotlib.pyplot as plt

plt.figure(figsize=(9, 12))

plt.subplot(2, 1, 1)
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.title('Original Image')

plt.subplot(2, 1, 2)
plt.plot(histrBlue, color='blue')
plt.plot(histrGreen, color='green')
plt.plot(histrRed, color='red')
plt.title('Histogram RGB')

plt.tight_layout()
plt.show()
```

```python
import cv2

img = cv2.imread("assets/Wenclair.jpeg")
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

face_cascade = cv2.CascadeClassifier("haarcascade_frontalface_default.xml")

# Deteksi wajah pada gambar
faces = face_cascade.detectMultiScale(
    gray,              # Gambar grayscale
    scaleFactor=1.1,   # Faktor untuk mengecilkan ukuran gambar
    minNeighbors=1     # Jumlah tetangga minimum untuk mempertahankan deteksi
)

# Menandai setiap wajah yang terdeteksi dengan kotak deteksi
for (x, y, w, h) in faces:
    # Menggambar kotak deteksi di sekitar wajah
    img = cv2.rectangle(
        img,            # Gambar asli
        (x, y),         # Koordinat kiri atas kotak
        (x + w, y + h), # Koordinat kanan bawah kotak
        (255, 255, 255),    # Warna kotak (hijau dalam format BGR)
        3               # Ketebalan garis kotak
    )
    # Memisahkan area wajah pada gambar grayscale dan berwarna
    roi_gray = gray[y:y+h, x:x+w]
    roi_color = img[y:y+h, x:x+w]

cv2.imshow('Hasil Deteksi Wajah', img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
cap = cv2.VideoCapture(0)
face_cascade = cv2.CascadeClassifier("haarcascade_frontalface_default.xml")

while cap.isOpened():
    ret, frame = cap.read()

    if ret:
        gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

        faces = face_cascade.detectMultiScale(
            gray,
            scaleFactor=1.3,
            minNeighbors=5
        )

        for (x, y, w, h) in faces:
            cv2.rectangle(
                frame,
                (x, y),
                (x + w, y + h),
                (0, 255, 0),
                2
            )

        cv2.imshow('Realtime Face Detection', frame)

        if cv2.waitKey(25) & 0xFF == ord('q'):
            break
    else:
        break

cap.release()
cv2.destroyAllWindows()
```
