```python
import cv2

# Read image
img = cv2.imread("assets/enid.png")

# Show image
cv2.imshow("Enid", img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import cv2

img = cv2.imread("assets/enid.png", cv2.IMREAD_GRAYSCALE)
cv2.imshow("Convert", img)
cv2.waitKey(0)
cv2.destroyAllWindows()

# Write image
cv2.imwrite('path_to_output.jpg', img)
```
