```python
import cv2
import matplotlib.pyplot as plt


img = cv2.imread("assets/enid.png")
image = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)


half = cv2.resize(image, (0, 0), fx = 0.5, fy = 0.5)
bigger = cv2.resize(image, (0,0), fx = 2, fy = 2)
stretch_near = cv2.resize(image, (1000, 500), 
               interpolation = cv2.INTER_LINEAR)


Titles =["Original", "Half", "Bigger", "Interpolation Nearest"]
images =[image, half, bigger, stretch_near]
count = 4

for i in range(count):
    plt.subplot(2, 2, i + 1)
    plt.title(Titles[i])
    plt.imshow(images[i])

plt.show()
```

```python
# importing libraries 
import cv2 
import numpy as np 

image = cv2.imread("assets/enid.png")

# Original Image
cv2.imshow('Original Image', image) 
cv2.waitKey(0) 

# Gaussian Blur 
Gaussian = cv2.GaussianBlur(image, (7, 7), 0) 
cv2.imshow('Gaussian Blurring', Gaussian) 
cv2.waitKey(0) 

# Median Blur 
median = cv2.medianBlur(image, 5) 
cv2.imshow('Median Blurring', median) 
cv2.waitKey(0) 

# Bilateral Blur 
bilateral = cv2.bilateralFilter(image, 9, 75, 75) 
cv2.imshow('Bilateral Blurring', bilateral) 
cv2.waitKey(0) 

cv2.destroyAllWindows() 
```

```python
import cv2

img = cv2.imread("assets/enid.png", cv2.IMREAD_COLOR)

img_rotate_90_CLK = cv2.rotate(img, cv2.ROTATE_90_CLOCKWISE)
img_rotate_90_CTR = cv2.rotate(img, cv2.ROTATE_90_COUNTERCLOCKWISE)
img_rotate_180 = cv2.rotate(img, cv2.ROTATE_180)

cv2.imshow('Normal', img)
cv2.waitKey(0)
cv2.imshow('Rotate 90 Clockwise', img_rotate_90_CLK)
cv2.waitKey(0)
cv2.imshow('Rotate 90 Counterclockwise', img_rotate_90_CTR)
cv2.waitKey(0)
cv2.imshow('Rotate 180', img_rotate_180)


cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import cv2

img = cv2.imread("assets/enid.png", cv2.IMREAD_COLOR)
img = cv2.resize(img,  (0, 0), fx=0.75, fy=0.75)


img_flip_h = cv2.flip(img, 1)
img_flip_v = cv2.flip(img, 0)
img_flip_vh = cv2.flip(img, -1)

cv2.imshow('No Flip', img)
cv2.waitKey(0)
cv2.imshow('Flip Horizontal', img_flip_h)
cv2.waitKey(0)
cv2.imshow('Flip Vertical', img_flip_v)
cv2.waitKey(0)
cv2.imshow('Flip Horizontal & Vertical', img_flip_vh)


cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import cv2

img = cv2.imread("assets/enid.png", cv2.IMREAD_COLOR)
img = cv2.resize(img,  (0, 0), fx=0.75, fy=0.75)

img_RGB2BGR = cv2.cvtColor(img, cv2.COLOR_RGB2BGR)
img_RGB2HSV = cv2.cvtColor(img, cv2.COLOR_RGB2HSV)
img_RGB2LAB = cv2.cvtColor(img, cv2.COLOR_RGB2LAB)

cv2.imshow('RGB', img)
cv2.waitKey(0)
cv2.imshow('BGR', img_RGB2BGR)
cv2.waitKey(0)
cv2.imshow('HSV', img_RGB2HSV)
cv2.waitKey(0)
cv2.imshow('LAB', img_RGB2LAB)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import matplotlib.pyplot as plt

plt.subplot(2, 2, 1)
plt.imshow(img)
plt.title('RGB')

plt.subplot(2, 2, 2)
plt.imshow(img_RGB2BGR)
plt.title('BGR')

plt.subplot(2, 2, 3)
plt.imshow(img_RGB2HSV)
plt.title('HSV')

plt.subplot(2, 2, 4)
plt.imshow(img_RGB2LAB)
plt.title('LAB')

plt.show()
```

```python
# importing libraries
import cv2

cap = cv2.VideoCapture(0)

if (cap.isOpened()== False):
    print("Error opening video file")

while(cap.isOpened()):
    ret, frame = cap.read()
    if ret == True:
        cv2.imshow('Frame', frame)
        if cv2.waitKey(25) & 0xFF == ord('q'):
            break
    else:
        break

cap.release()
cv2.destroyAllWindows()
```
